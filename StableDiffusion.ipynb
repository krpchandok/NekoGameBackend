```python
from diffusers import DiffusionPipeline, StableDiffusionXLImg2ImgPipeline
import torch
import os

# ✅ Make sure output folders exist
os.makedirs("images", exist_ok=True)
os.makedirs("images_refined", exist_ok=True)

# ✅ Load base model
model = "stabilityai/stable-diffusion-xl-base-1.0"
pipe = DiffusionPipeline.from_pretrained(
    model,
    torch_dtype=torch.float16,
    variant="fp16",  # ensure you're loading correct weights
)
pipe.to("cuda")

# ✅ Load LoRA fine-tuned weights
pipe.load_lora_weights("model/", weight_name="pytorch_lora_weights.safetensors")

# ✅ Load refiner
refiner = StableDiffusionXLImg2ImgPipeline.from_pretrained(
    "stabilityai/stable-diffusion-xl-refiner-1.0",
    torch_dtype=torch.float16,
    variant="fp16",
)
refiner.to("cuda")

# ✅ Prompts
prompt = "a really hairy creature"
negative_prompt = "a human being"

# ✅ Generate and refine images
for seed in range(10):
    generator = torch.Generator("cuda").manual_seed(seed)
    
    # Base generation
    output = pipe(
        prompt=prompt,
        negative_prompt=negative_prompt,
        guidance_scale=15,
        num_inference_steps=25,
        generator=generator,
    )
    image = output.images[0]
    image.save(f"images/{seed}.png")

    # Refine
    refined_output = refiner(
        prompt=prompt,
        image=image,
        num_inference_steps=25,
        guidance_scale=15,
        generator=generator,
    )
    refined_image = refined_output.images[0]
    refined_image.save(f"images_refined/{seed}.png")
```
